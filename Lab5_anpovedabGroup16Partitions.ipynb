### Write a code that produces the the **Bell numbers**. It count the possible partitions of a set.

**Hint**:
. Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number)

```python
def bell(n):
    triangle = [1]

    for _ in range(n):
        pivot = triangle[0]
        triangle[0] = triangle[-1]
        for i in range(1, len(triangle)):
            tmp  = triangle[i]
            triangle[i] = pivot + triangle[i-1]
            pivot = tmp
        triangle.append(pivot + triangle[-1])

    return triangle[0]

for i in range(16):
    print(f'{i}: {bell(i)}')
```

### Write a code that that produces  list of all partitions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for which the sum of elements of the partition is the same constant for all.

```python
def get_partitions(numbers, index, subset, partitions):
    if index == len(numbers):
        partitions.append([list(s) for s in subset])
        return

    for i in range(len(subset)):
        subset[i].append(numbers[index])
        get_partitions(numbers, index+1, subset, partitions)
        subset[i].pop()

    subset.append([numbers[index]])
    get_partitions(numbers, index+1, subset, partitions)
    subset.pop()


n = 7
partitions = []
numbers = list(range(1, n+1))
get_partitions(numbers, 0, [], partitions)

print('Partitions')
for partition in partitions:
    print(partition)
```

### Write a code that produces the partitions with all the parts having the same sum value, identify the formula for $n < 16$.

```python
def partitions_equal_sum(partitions):
    print('Partitions with an equal sum in all its subsets')
    for partitions in partitions:
        first_sum = sum(partitions[0])
        equal = True
        for subset in partitions:
            sum_elements = sum(subset)
            if sum_elements != first_sum:
                equal = False
                break
        if equal:
            print(f'Elements: {partitions}. Sum: {first_sum}')

partitions_equal_sum(partitions)
```
